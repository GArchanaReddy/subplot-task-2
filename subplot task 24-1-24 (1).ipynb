# subplot :
subplot method provides a way to plot multiple plot in a single figure.
the argements are rows ,columns,index
the uses of subplot is allows you to insert multiple plots on a grid within in a single figure. the basis form ofn the subplot()command takes in three inputs.nrows,ncols.linearindex,. the  first two arguments define the number of rows and columns that will be included in the grid

```python
import matplotlib.pyplot as plt, numpy as np , pandas as pd, seaborn as sns
```

```python
np.random.seed(42)

# Create a 3x3 grid of plots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

# Violin Plot in the first row, first column
G = gridspec.GridSpec(3, 3)

# Plotting the Violin Plot in the first row
ax = plt.subplot(G[0, :])
sns.violinplot(d['sepal_length']).set_title('Violin Plot for Sepal Length')

#sns.violinplot(x=df['sepal_length'], ax=axes[0,3])
#axes[0, 0].set_title('Violin Plot')

# Histogram Plot in the first row, second column
plt.subplot(3, 3, 4)
sns.histplot(d['sepal_width']).set_title('Histogram')

# Density Plot in the first row, third column
plt.subplot(3, 3, 5)
sns.kdeplot(d['petal_length']).set_title('Density Plot')
axes[1, 1].axvline(np.mean(d['petal_length']), color='r', linestyle='--', label='Mean')
axes[1, 1].axvline(np.median(d['petal_length']), color='g', linestyle='--', label='Median')
axes[1, 1].axvline(np.mean(d['petal_length']) + np.std(d['petal_length']), color='b', linestyle='--')
axes[1, 1].axvline(np.mean(d['petal_length']) - np.std(d['petal_length']), color='b', linestyle='--')
axes[1, 1].axvline(np.mean(d['petal_length']) + 2 * np.std(d['petal_length']), color='m', linestyle='--')
axes[1, 1].axvline(np.mean(d['petal_length']) - 2 * np.std(d['petal_length']), color='m', linestyle='--')
axes[1, 1].axvline(np.mean(d['petal_length']) + 3 * np.std(d['petal_length']), color='y', linestyle='--')
axes[1, 1].axvline(np.mean(d['petal_length']) - 3 * np.std(d['petal_length']), color='y', linestyle='--')
axes[1, 1].legend()


# Scatter Plot in the second row, third column
plt.subplot(3, 3, 6)
axes[1, 2].scatter(d['petal_width'], d['sepal_width'])
axes[1, 2].set_title('Scatter Plot')

# Pie Plot in the second row, first column
data_pie = d['species'].value_counts()
plt.subplot(3, 3, 7)
plt.pie(data_pie, labels=data_pie.index, autopct='%1.1f%%', startangle=90)
plt.title('Exploded Pie Plot')

# Count Plot in the second row, second column
plt.subplot(3, 3, 8)
sns.countplot(x=d['species'], order=d['species'].value_counts().index, palette='viridis')
plt.title('Count Plot')

# Butterfly Plot in the second row, third column
plt.subplot(3, 3, 9)
data_left = np.random.randint(1, 10, size=(5,))
data_right = np.random.randint(1, 10, size=(5,))
plt.title('Butterfly Plot')

plt.barh(np.arange(len(data_left)), data_left, color='orange', label='Left Wing')
plt.barh(np.arange(len(data_right)), -data_right, color='green', label='Right Wing')

plt.yticks(np.arange(len(data_left)), ['A', 'B', 'C', 'D', 'E'])
plt.xlabel('Values')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
